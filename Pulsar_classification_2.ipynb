# Pulsar Star Classification

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('data/train.csv').drop('id', axis=1)
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class', axis=1), df['Class'], test_size=0.2, random_state=42)
```

```python
df.head()
```

```python
df['EK*SD'] = df['EK'] * df['SD']
#df['SD/EK'] = df['SD'] / df['EK']
df['Skewness*SD'] = df['Skewness'] * df['SD']
#df['Mean_DMSNR_Curve/SD_DMSNR_Curve'] = df['Mean_DMSNR_Curve'] / df['SD_DMSNR_Curve']
#df['SD_DMSNR_Curve/EK_DMSNR_Curve'] = df['SD_DMSNR_Curve'] / df['EK_DMSNR_Curve']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class', axis=1), df['Class'], test_size=0.2, random_state=42)
X,y = df.drop('Class', axis=1), df['Class']
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score

cat = CatBoostClassifier(random_state=42,loss_function='Logloss',max_depth=2,verbose=50,iterations=1000,eval_metric='Logloss')
scores = cross_val_score(cat, X_train, y_train, cv=10, scoring='neg_log_loss', n_jobs=-1)
print('Log Loss: ', -scores.mean())
```

```python
from catboost import CatBoostClassifier
cat = CatBoostClassifier(random_state=42,loss_function='Logloss',max_depth=2,verbose=50,iterations=25000,eval_metric='Logloss',learning_rate=0.005,l2_leaf_reg=2)
```

```python
cat.fit(X_train, y_train,eval_set=(X_test, y_test),use_best_model=True,plot=True)
```

```python
#grid search on catboost
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
cat = CatBoostClassifier(random_state=42,loss_function='Logloss',verbose=50,iterations=1000,eval_metric='Logloss')
grid = {'learning_rate': [0.01,0.1,0.5,1],
        'l2_leaf_reg': [1,3,5,7,9],
        'max_depth': [2,4,6,8]}
grid_search_result = cat.grid_search(grid, X=X_train, y=y_train,plot=True)
```

```python
#print the best parameters
print('Best params:' + str(grid_search_result['params']))
```

```python
#f1 score
from sklearn.metrics import f1_score
y_pred = cat.predict(X_test)
print('F1 score: ', f1_score(y_test, y_pred))
y_pred = cat.predict(X_train)
print('F1 score: ', f1_score(y_train, y_pred))
# Log Loss score
from sklearn.metrics import log_loss
y_pred = cat.predict_proba(X_test)[:,1]
print('Log Loss: ', log_loss(y_test, y_pred))
y_pred = cat.predict_proba(X_train)[:,1]
print('Log Loss: ', log_loss(y_train, y_pred))
```

```python
cat.fit(X, y, verbose=50)
```

```python
df_test = pd.read_csv('data/test.csv')
df_test['EK*SD'] = df_test['EK'] * df_test['SD']
#df_test['SD/EK'] = df_test['SD'] / df_test['EK']
#df_test['SD_DMSNR_Curve/EK_DMSNR_Curve'] = df_test['SD_DMSNR_Curve'] / df_test['EK_DMSNR_Curve']
df_test['Skewness*SD'] = df_test['Skewness'] * df_test['SD']
#df_test['Mean_DMSNR_Curve/SD_DMSNR_Curve'] = df_test['Mean_DMSNR_Curve'] / df_test['SD_DMSNR_Curve']
y_pred = cat.predict_proba(df_test.drop(['id'], axis=1))[:,1]
df_pred = pd.DataFrame({'id':df_test['id'], 'Class':y_pred})
df_pred.head()
```
